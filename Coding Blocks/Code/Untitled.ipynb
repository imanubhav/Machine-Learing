```python
#plot simple ranges
plt.style.use('seaborn-darkgrid')
a=np.array(range(0,100,2))
#print(a)
b=np.array(range(0,100,2))
#print(b)

plt.figure(0)
plt.title("Line")
plt.grid("on")
plt.xlabel("year")
plt.xlim(0,100)
plt.ylabel("sales")
plt.ylim(0,100)
plt.plot(a,b,'r',label="2018")
plt.scatter(a,2*a, c='g',label="2017", marker="*")
plt.legend()
plt.show()
```

```python
#load the movie meta-data dataset
df=pd.read_csv("movie_metadata.csv")
#print(df.head(5))
#df.info
```

```python
#get the values as numpy array and check shape
values=df.values
print(values.shape, type(values))
```

```python
#print all columns
print(df.columns)
```

```python
#Print a few titles

titles=df.get('movie_title')
titles.head()
lens=[]
for ix in titles.values:
    lens.append(len(ix))
#print(len(lens))
#print(type(lens))
```

```python
freq={}

for ix in lens:
    curr=freq.get(ix,0)
    freq[ix]=curr+1 
```

```python
plt.plot(freq.keys(), freq.values())
```
